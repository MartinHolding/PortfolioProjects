<h3> Spread of species by conservation status </h3>
First, I'm going to look at the spread of species across the parks, listed by their conservation status.


```python
#First I need a new df with my animal name, category, conservation_status and park_name
#This can be easily achieived using a left join. 
new_df = species_data.merge(obs_data, left_on='scientific_name', right_on='scientific_name')
new_df
```

```python
# Plot the number of each type of life by it's conservation status in each park
park_names = new_df.park_name.unique()
con_status = new_df.conservation_status.unique()
y_data
for park in park_names:
    y_data.append(Counter(new_df[new_df.park_name == park].conservation_status))#count number of occurences in each park

y_data[0][1].items()

#y_data_items = y_data.items

#y_data_items
#plt.bar(new_df.park_name, new_df.conservation_status)
#plt.show
```
